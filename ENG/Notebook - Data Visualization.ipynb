# Data Visualization Basics

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

## Creating Data

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

## First Graphic


```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

## Axes and Legend

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend(('sin', 'cos'))
```

## Subplots

```python
plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.xlabel('X-Axis')
plt.plot(x, np.sin(x))
plt.ylabel('Y-Axis')


plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

## Design Changes

```python
fig2 = plt.figure(figsize=(10, 4))  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.xlabel('X-Axis')
plt.plot(x, np.sin(x))
plt.ylabel('Y-Axis')


# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), color='red', linestyle='dashed', linewidth=4);
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

## Saving Image

```python
fig2.savefig('first_image.png')
```

# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

To show the line plots, let’s first import the famous iris data set.

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['petal_length'].plot()
```

```python
sns.lineplot(data=iris, y='petal_length', x=iris.index)
```

```python
sns.lineplot(data=iris, y='petal_length', x=iris.index, hue='species')
```

## Scatter Plot

```python
fig, ax = plt.subplots()

ax.scatter(iris['sepal_length'], iris['sepal_width'])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

## Bar Plot

```python
iris.groupby(by='species').mean()['petal_length']
```

```python
data = iris.groupby(by='species').mean()['petal_length']
ax = sns.barplot(x=data.index, y=data.values)


ax.set_title('Petal Length Means') 
ax.set_xlabel('Species') 
ax.set_ylabel('Mean')
```

## Histogram

```python
sns.histplot(data=iris, x='petal_length')
```

```python
sns.histplot(data=iris, y='petal_length')
```

```python
sns.histplot(data=iris, x='petal_length', hue='species')
```

```python
sns.histplot(data=iris, x='petal_length', hue='species', bins=20)
```

## Heatmap

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
sns.heatmap(iris.corr(), annot=True)
```

## Box Plot

```python
sns.boxplot(y='sepal_length', x='species', data=iris)
```
