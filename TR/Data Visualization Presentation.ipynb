# Veri Gorsellestirme Temel Bilgiler

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

## Data Olusturalım

```python
x = np.linspace(0, 10, 100)
```

## Ilk Grafik

```python
fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

## Eksenler ve Lejant

```python
fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')

plt.legend(('sin', 'cos'))
```

## Subplotlar

```python
fig2 = plt.figure(figsize=(10, 4))  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.xlabel('x ekseni')
plt.plot(x, np.sin(x))
plt.ylabel('y ekseni')


# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
```

## Tasarımsal Değişiklikler

```python
fig2 = plt.figure(figsize=(10, 4))  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.xlabel('x ekseni')
plt.plot(x, np.sin(x))
plt.ylabel('y ekseni')


# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), color='red', linestyle='dashed', linewidth=4);
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
```
